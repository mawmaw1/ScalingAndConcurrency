```python
%%sql 
SELECT drivers.forename, drivers.surname, drivers.driverid, count(*) AS wins
FROM drivers
JOIN results USING (driverid)
WHERE results.position = 1
GROUP BY driverid
ORDER BY count(*)
DESC;
```

## 2)

```python
%%sql 
SELECT driverref, constructors.name, count(*)
FROM drivers 
JOIN results USING (driverid)
JOIN constructors USING (constructorid)
GROUP BY driverref, constructors.name
ORDER BY count(*)
DESC;
```

## 3)

```python
%%sql
SELECT driverref, constructors.name, count(*) AS driveCount
FROM drivers
JOIN results USING (driverid)
JOIN constructors USING (constructorid)
WHERE results.statusid = 1
GROUP BY driverref, constructors.name
ORDER BY count(*)
DESC;
```

```python
%%sql CREATE VIEW drivenCounter AS
SELECT driverref, constructors.name, count(*) AS driveCount
FROM drivers
JOIN results USING (driverid)
JOIN constructors USING (constructorid)
WHERE results.statusid = 1
GROUP BY driverref, constructors.name
ORDER BY count(*)
DESC;
```

## 4)

```python
%%sql CREATE VIEW pitstopsumse AS
SELECT pitstops.raceid, pitstops.driverid, results.constructorid, sum(pitstops.milliseconds) AS pitstopsum
FROM pitstops
INNER JOIN results
ON pitstops.driverid = results.driverid AND pitstops.raceid = results.raceid
GROUP BY pitstops.raceid, pitstops.driverid, results.constructorid
ORDER BY pitstops.raceid, pitstops.driverid
```

## 5)

```python
%%sql 
SELECT driverref, constructors.name, sum(pitstopq.pitstopsum)
FROM drivers
JOIN results USING (driverid)
JOIN constructors USING (constructorid)
INNER JOIN (SELECT driverid, pitstopsum,constructorid FROM pitstopsumse) AS pitstopq
ON (drivers.driverid = pitstopq.driverid AND constructors.constructorid = pitstopq.constructorid)
WHERE results.statusid = 1
GROUP BY driverref, constructors.name
ORDER BY driverref
DESC;
```

## 6 )

```python
%%sql 
SELECT drivers.driverid,driverref, constructors.name, sum(pitstopq.pitstopsum),
ROUND( sum(pitstopq.pitstopsum) / count(drivers.driverid), 2) AS avg
FROM drivers
JOIN results USING (driverid)
JOIN constructors USING (constructorid)
INNER JOIN (SELECT driverid, pitstopsum,constructorid FROM pitstopsumse) AS pitstopq
ON (drivers.driverid = pitstopq.driverid AND constructors.constructorid = pitstopq.constructorid)
WHERE results.statusid = 1
GROUP BY driverref, constructors.name,drivers.driverid
ORDER BY avg
ASC;
```
